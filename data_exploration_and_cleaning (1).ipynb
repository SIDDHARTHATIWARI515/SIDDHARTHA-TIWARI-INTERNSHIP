# Python Basics & Data Exploration/Cleaning with Pandas
This notebook demonstrates the foundational steps of data loading, exploration, handling missing values, standard data operations, removing duplicates, creating derived columns, and saving a cleaned dataset.

### Step 1: Load a CSV dataset into a Pandas DataFrame

```python
import pandas as pd
import numpy as np

df = pd.read_csv('Combined_dataset.csv')
print("Dataset loaded successfully!")
```

### Step 2: Explore Data (head/tail, shape, columns, data types)

```python

print("--- First 5 Rows ---")
print(df.head())


print("\n--- Last 5 Rows ---")
print(df.tail())


print(f"\nShape of the dataset: {df.shape}")


print(f"\nColumns: {list(df.columns)}")


print("\n--- Data Types ---")
print(df.dtypes)
```

### Step 3: Handle Missing Values (Identify and Fill/Drop)

```python
print("--- Missing Values Per Column ---")
print(df.isnull().sum())

df['discount'] = df['discount'].fillna(0)

text_cols_with_nan = ['what_customers_said', 'seller_name', 'videos', 'seller_information', 'variations']
for col in text_cols_with_nan:
    df[col] = df[col].fillna('Unknown')

print("\nMissing values handled!")
print(f"Remaining total missing values: {df.isnull().sum().sum()}")
```

### Step 4: Perform Basic Operations (Filter rows, select columns)

```python
highly_rated_df = df[df['rating'] >= 4.0]
print(f"Number of products with rating >= 4.0: {len(highly_rated_df)}")

selected_columns_df = df[['product_id', 'title', 'initial_price', 'discount', 'rating', 'category']]
print("\n--- Selected Columns Subset ---")
print(selected_columns_df.head())
```

### Step 5: Remove Duplicates

```python
duplicate_count = df.duplicated(subset=['product_id']).sum()
print(f"Number of duplicate products: {duplicate_count}")


df = df.drop_duplicates(subset=['product_id'], keep='first')
print(f"Shape after removing duplicates: {df.shape}")
```

### Step 6: Create a Derived Column (total_amount = price * quantity)
Since the original dataset does not contain a `quantity` column, we will simulate a `quantity` column representing the number of items ordered/available, and then calculate the `total_amount` using `initial_price`.

```python
np.random.seed(42)
df['quantity'] = np.random.randint(1, 6, size=len(df))

df['total_amount'] = df['initial_price'] * df['quantity']

print("--- Derived Columns Added ---")
print(df[['title', 'initial_price', 'quantity', 'total_amount']].head())
```

### Step 7: Save the Cleaned Dataset as a New CSV File

```python
df.to_csv('cleaned_dataset.csv', index=False)
print("Cleaned dataset saved as 'cleaned_dataset.csv' successfully!")
```
