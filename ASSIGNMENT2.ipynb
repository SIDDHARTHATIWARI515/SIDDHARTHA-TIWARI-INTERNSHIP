```python
import pandas as pd
import sqlite3
df = pd.read_csv("Sample - Superstore.csv", encoding="latin1")
df.columns = [c.replace(' ', '_').replace('-', '_') for c in df.columns]
conn = sqlite3.connect(':memory:')
df.to_sql('superstore', conn, index=False, if_exists='replace')
print("Success! The 'superstore' table is created and loaded into your SQL database.")
```

```python
query_schema = "PRAGMA table_info(superstore);"
pd.read_sql_query(query_schema, conn)
```

```python
query_sample = "SELECT * FROM superstore LIMIT 5;"
pd.read_sql_query(query_sample, conn)
```

```python
query_filters = """
SELECT Order_ID, State, Product_Name, Sales, Profit
FROM superstore
WHERE Category = 'Technology'
  AND Sales > 500
ORDER BY Sales DESC
LIMIT 3;
"""
pd.read_sql_query(query_filters, conn)
```

```python
query_aggregations = """
SELECT
    Category,
    ROUND(SUM(Sales), 2) AS Total_Sales,
    SUM(Quantity) AS Total_Units_Sold,
    ROUND(AVG(Profit), 2) AS Avg_Profit_Per_Unit
FROM superstore
GROUP BY Category
ORDER BY Total_Sales DESC;
"""
pd.read_sql_query(query_aggregations, conn)
```

```python
query_top_products = """
SELECT Product_Name, Category, ROUND(SUM(Sales), 2) AS Total_Revenue
FROM superstore
GROUP BY Product_Name, Category
ORDER BY Total_Revenue DESC
LIMIT 3;
"""
pd.read_sql_query(query_top_products, conn)
```
